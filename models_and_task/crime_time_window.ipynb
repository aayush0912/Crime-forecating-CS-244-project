### Importing packages and reading data form CSV file

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')

data_dir='E:/pro_cs244/crime_data.csv'

df0 = pd.read_csv(data_dir)
df0.head(5)
print(len(df0))
```

```python
df=df0.dropna()
```

```python
df.info()
```

```python
# Extracting time_window from Date column
def make_window(row):
    date, time, half = row.split()
    h = int(time[0:2])
    if(half == 'AM'):
        if(h <= 6):
            return 'night'
        else:
            return 'morning'
    if(half == 'PM'):
        if(h <= 6):
            return 'afternoon'
        else:
            return 'evening'
        
```

```python
df['time_window'] = df.apply(lambda row: make_window(row['Date']), axis=1)
```

```python
df['time_window']
```

```python
print(df.groupby('Primary Type').size())
```

```python
df['Primary Type'].value_counts().plot.bar()
plt.title("Crimes_types")
plt.show()
```

### time window for each type of crime

```python
color = (0.2, 0.4, 0.6, 0.8)

df[df.time_window == 'morning']['Primary Type'].value_counts(normalize=False).plot.bar(color=color)
plt.title("Crimes commited in morning")
plt.show()

df[df.time_window == 'night']['Primary Type'].value_counts(normalize=False).plot.bar(color=color)
plt.title("Crimes commited in night")
plt.show()

df[df.time_window == 'afternoon']['Primary Type'].value_counts(normalize=False).plot.bar(color=color)
plt.title("Crimes commited in afternoon")
plt.show()

df[df.time_window == 'evening']['Primary Type'].value_counts(normalize=False).plot.bar(color=color)
plt.title("Crimes commited in evening")
plt.show()
```
